```python
#creating dataframe from dictionary
data = {
    "day": ["25/12/2023", "14/07/2023", "21/04/2023", "20/01/2023"],
    "event": ["Sunny", "Rain", "Snow", "Rain"],
    "windspeed": [2, 6, 7, 8]
}
df = pd.DataFrame(data)
df
```

```python
#creating dataframe from list of dictionary
data = [
    {"date": "25/12/2023", "Event": "Rainy", "Windspeed": 4},
    {"date": "21/04/2023", "Event": "Sunny", "Windspeed": 6},
    {"date": "15/07/2023", "Event": "Rainy", "Windspeed": 3},
]
df = pd.DataFrame(data)
df
```

```python
#creating dataframe from tuples
data = (
    ("25/12/2023", "Rainy", 4),
    ("21/04/2023", "Sunny", 6),
    ("15/07/2023", "Rainy", 3),
)
df = pd.DataFrame(data, columns=["Date", "Event", "Windspeed"])
df
```
