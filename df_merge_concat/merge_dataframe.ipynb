```python
import pandas as pd
pl_goal = pd.DataFrame({
    "players" : ["haaland", "salah", "solanke"],
    "goal" : [14, 14, 12]
})
pl_goal
```

```python
pl_apps = pd.DataFrame(
{
    "players" : ["haaland", "solanke", "salah"],
    "apps" : [16, 25, 20]
})
pl_apps
```

```python
#performs inner join
new_df = pd.merge(pl_goal, pl_apps, on="players")
new_df
```

```python
pl_goal = pd.DataFrame({
    "players" : ["haaland", "enzo", "toney"],
    "goal" : [14, 4, 2]
}, index=[0, 1,2])
pl_goal
```

```python
pl_apps = pd.DataFrame(
{
    "players" : ["haaland", "solanke", "nunez"],
    "apps" : [16, 25, 16]
})
pl_apps
```

```python
#only haaland present in both dataframe
#intersection
inner = pd.merge(pl_goal, pl_apps, on="players")
inner
```

```python
#database joins like outer, left, right join can also be performed using merge
#indicator -> tells the dataframe from the entry came from
outer = pd.merge(pl_goal, pl_apps, on="players", how="outer", indicator=True)
outer
```

```python
pl = pd.DataFrame({
    "players" : ["haaland", "salah", "solanke"],
    "goals" : [14, 14, 12],
    "apps" : [16, 20, 20]
})
pl
```

```python
laliga = pd.DataFrame({
    "players" : ["dovbyk", "bellingham", "mayoral"],
    "goals" : [14, 14, 14],
    "apps" : [21, 18, 21]
})
laliga
```

```python
#merging la liga and pl dataframe together
new_df = pd.merge(pl, laliga, how="outer")
new_df
```

```python
season_2024 = pd.DataFrame({
    "players" : ["haaland", "salah", "solanke"],
    "goals" : [14, 14, 12],
    "apps" : [16, 20, 20]
})
season_2024
```

```python
season_2023 = pd.DataFrame({
    "players" : ["haaland", "salah", "kane"],
    "goals" : [38, 32, 36],
    "apps" : [32, 27, 24]
})
season_2023
```

```python
#both df contains salah as well as haaland
#merging withot suffixes
new_df = pd.merge(season_2023, season_2024, on="players")
new_df
```

```python
#to give meaningful suffix when the dfs are merged
new_df = pd.merge(season_2023, season_2024, on="players", suffixes=('_2023', '_2024'))
new_df
```
