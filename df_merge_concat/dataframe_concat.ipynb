```python
#concatenating dataframes
import pandas as pd

pl = pd.DataFrame({
    "players" : ["haaland", "salah", "solanke"],
    "goals" : [14, 14, 12],
    "apps" : [16, 20, 20]
})

laliga = pd.DataFrame({
    "players" : ["dovbyk", "bellingham", "mayoral"],
    "goals" : [14, 14, 14],
    "apps" : [21, 18, 21]
})
```

```python
#concatenates two dataframe
#ignroes the index from original dataframe
df = pd.concat([pl, laliga], ignore_index=True)
df
```

```python
#indexing each df with key
#doesnt wotk with ignore_index being true
df = pd.concat([pl, laliga], keys=["pl", "laliga"])
df
```

```python
#returns only the laliga players
df.loc["laliga"]
```

```python
#concatenating column-wise
#indexed to preserve the order
pl_goal = pd.DataFrame({
    "players" : ["haaland", "salah", "solanke"],
    "goal" : [14, 14, 12]
}, index=[0, 1,2])
pl_goal
```

```python
pl_apps = pd.DataFrame(
{
    "players" : ["haaland", "solanke", "salah"],
    "apps" : [16, 25, 20]
}, index=[0,2,1])
pl_apps
```

```python
df2 = pd.concat([pl_goal, pl_apps], axis=1)
df2
```

```python
#series returns data in a column wise manner
s = pd.Series(df.apply(lambda d : d.goals / d.apps, axis=1), name="goals_per_match")
s
```

```python
new_df = pd.concat([df, s], axis=1)
new_df
```
